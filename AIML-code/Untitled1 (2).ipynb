1.Uninformed Search Techniques: -BFS -DFS -IDS

```python
# Breadth First Search
L = [
    ('a', ['b', 'c']),
    ('b', ['d', 'e']),
    ('c', ['f']),
    ('d', ['g']),
    ('e', ['h']),
    ('f', ['i']),
    ('g', []),
    ('h', []),
    ('i', ['j']),
    ('j', ['k']),
    ('k', [])
]

graph = {node: neighbors for node, neighbors in L}

start = 'a'
goal = 'k'
queue = [[start]]
visited = []

while len(queue) > 0:
    path = queue.pop(0)
    node = path[-1]

    if node in visited:
        continue

    visited.append(node)

    if node == goal:
        print(path)
        break

    for neighbor in graph[node]:
        new_path = path + [neighbor]
        queue.append(new_path)

```

```python
#Iterative Deepening Search
L = [
    ('a', ['b', 'c']),
    ('b', ['d', 'e']),
    ('c', ['f']),
    ('d', ['g']),
    ('e', ['h']),
    ('f', ['i']),
    ('g', []),
    ('h', []),
    ('i', ['j']),
    ('j', ['k']),
    ('k', [])
]

graph = {node: neighbors for node, neighbors in L}

start = 'a'
goal = 'k'
depth_limit = 4
found = False


for limit in range(depth_limit + 1):
    stack = [[start]]
    visited = []

    while len(stack) > 0:
        path = stack.pop()
        node = path[-1]

        if node in visited:
            continue

        visited.append(node)

        if node == goal:
            print(path)
            found = True
            break

        if len(path) - 1 < limit:
            for neighbor in graph[node]:
                new_path = path + [neighbor]
                stack.append(new_path)

    if found:
        break

```

```python
#Depth First Search
L = [
    ('a', ['b', 'c']),
    ('b', ['d', 'e']),
    ('c', ['f']),
    ('d', ['g']),
    ('e', ['h']),
    ('f', ['i']),
    ('g', []),
    ('h', []),
    ('i', ['j']),
    ('j', ['k']),
    ('k', [])
]

graph = {node: neighbors for node, neighbors in L}

start = 'a'
goal = 'k'
stack = [[start]]
visited = []

while len(stack) > 0:
    path = stack.pop()
    node = path[-1]

    if node in visited:
        continue

    visited.append(node)

    if node == goal:
        print(path)
        break

    for neighbor in graph[node]:
        new_path = path + [neighbor]
        stack.append(new_path)

```


2.Informed Search Techniques: -A* search technique -Best first search

```python
#A* Search :
L = [
    ('a', ['b', 'c'], 6),
    ('b', ['d', 'e'], 5),
    ('c', ['f'], 4),
    ('d', ['g'], 3),
    ('e', ['h'], 2),
    ('f', ['i'], 1),
    ('g', [], 4),
    ('h', [], 3),
    ('i', ['j'], 0),
    ('j', ['k'], 2),
    ('k', [], 0)
]

graph = {node: neighbors for node, neighbors, _ in L}
heuristics = {node: h for node, _, h in L}

start = 'a'
goal = 'k'
open_list = [[start, 0 + heuristics[start]]]
visited = []

while len(open_list) > 0:
    open_list.sort(key=lambda x: x[1])
    path = open_list.pop(0)[0]
    node = path[-1]

    if node in visited:
        continue

    visited.append(node)

    if node == goal:
        print(path)
        break

    for neighbor in graph[node]:
        g_cost = len(path)
        f_cost = g_cost + heuristics[neighbor]
        new_path = path + neighbor
        open_list.append([new_path, f_cost])
```

```python
#Best-First Search:
L = [
    ('a', ['b', 'c'], 6),
    ('b', ['d', 'e'], 5),
    ('c', ['f'], 4),
    ('d', ['g'], 3),
    ('e', ['h'], 2),
    ('f', ['i'], 1),
    ('g', [], 4),
    ('h', [], 3),
    ('i', ['j'], 0),
    ('j', ['k'], 2),
    ('k', [], 0)
]

graph = {node: neighbors for node, neighbors, _ in L}
heuristics = {node: h for node, _, h in L}

start = 'a'
goal = 'k'
open_list = [[start, heuristics[start]]]
visited = []

while len(open_list) > 0:
    open_list.sort(key=lambda x: x[1])
    path = open_list.pop(0)[0]
    node = path[-1]

    if node in visited:
        continue

    visited.append(node)

    if node == goal:
        print(path)
        break

    for neighbor in graph[node]:
        new_path = path + neighbor
        open_list.append([new_path, heuristics[neighbor]])


```

3.Adversal Search Techniques:
-(min-max) -(alpha-beta)


```python
#min-max technique
decision_tree = {
    'score': 0,
    'options': [
        {'score': 0, 'options': [
            {'score': 10, 'options': []},
            {'score': 20, 'options': []}
        ]},
        {'score': 0, 'options': [
            {'score': 30, 'options': []},
            {'score': 40, 'options': []}
        ]}
    ]
}

max_depth = 3
maximize = True

node_stack = [(decision_tree, max_depth, maximize)]
optimal_score = float('-inf') if maximize else float('inf')

while node_stack:
    current_node, depth_remaining, is_maximizing = node_stack.pop()

    if depth_remaining == 0 or not current_node['options']:
        if is_maximizing:
            optimal_score = max(optimal_score, current_node['score'])
        else:
            optimal_score = min(optimal_score, current_node['score'])
    else:
        for option in current_node['options']:
            node_stack.append((option, depth_remaining - 1, not is_maximizing))

print("Optimal score for minimizing player:", optimal_score)
```

```python
#Alpha-Beta technique
graph = {
    'cost': 0,
    'children': [
        {'cost': 0, 'children': [
            {'cost': 3, 'children': []},
            {'cost': 8, 'children': []}
        ]},
        {'cost': 0, 'children': [
            {'cost': 1, 'children': []},
            {'cost': 6, 'children': []}
        ]}
    ]
}

depth = 3
is_minimizing = True

alpha = float('-inf')
beta = float('inf')
best_cost = float('-inf') if not is_minimizing else float('inf')

stack = [(graph, depth, is_minimizing, alpha, beta)]

while stack:
    node, current_depth, is_min, alpha, beta = stack.pop()

    if current_depth == 0 or not node['children']:

        cost = node['cost']
        if is_min:
            best_cost = min(best_cost, cost)
        else:
            best_cost = max(best_cost, cost)
    else:
        if is_min:
            min_eval = float('inf')
            for child in node['children']:
                stack.append((child, current_depth - 1, False, alpha, beta))


                if min_eval <= alpha:
                    break
                min_eval = min(min_eval, best_cost)
                beta = min(beta, min_eval)
            best_cost = min_eval
        else:
            max_eval = float('-inf')
            for child in node['children']:
                stack.append((child, current_depth - 1, True, alpha, beta))

                # Update alpha and check for pruning
                if max_eval >= beta:
                    break
                max_eval = max(max_eval, best_cost)
                alpha = max(alpha, max_eval)
            best_cost = max_eval

print("Optimal cost for maximizing player:", best_cost)
```
